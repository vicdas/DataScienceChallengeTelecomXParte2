#📌 Extracción

```python
import pandas as pd # Manejar biblioteca pandas
import requests # Manejar biblioteca requests

url = 'https://raw.githubusercontent.com/vicdas/DataScienceChallengeTelecomX/refs/heads/main/TelecomX_Data.json'

response = requests.get(url)
data = response.json()
```

```python
data
```

```python
# Normalizar (aplanar) estructuras anidadas
datos_telecomx = pd.json_normalize(data)
datos_telecomx.head()
```

```python
type(datos_telecomx)
```

#🔧 Transformación

```python
datos_telecomx.info()
```

```python
datos_telecomx.dtypes
```

```python
import numpy as np

# Reemplazar espacios con NaN antes de convertir a float
datos_telecomx['account.Charges.Total'] = datos_telecomx['account.Charges.Total'].replace(' ', np.nan)

# Usando pd.to_numeric con errors='coerce' para manejar otro valor no numérico
datos_telecomx['account.Charges.Total'] = pd.to_numeric(datos_telecomx['account.Charges.Total'], errors='coerce')

# Ahora a convertirlo en un float64
datos_telecomx['account.Charges.Total'] = datos_telecomx['account.Charges.Total'].astype(np.float64)
```

### Paso 1: Verificar valores ausentes

```python
# Cantidad de valores nulos por columna
print(datos_telecomx.isnull().sum())
```

```python
# Crear una tabla con el conteo y porcentaje de valores nulos
missing_data = datos_telecomx.isnull().sum().to_frame(name='Nulos')
missing_data['% Nulos'] = 100 * missing_data['Nulos'] / len(datos_telecomx)
missing_data = missing_data[missing_data['Nulos'] > 0].sort_values(by='Nulos', ascending=False)

# Mostrar la tabla
print(missing_data)
```

```python
# Usar la media para llenar con valores los datos faltantes
datos_telecomx['account.Charges.Total'] = datos_telecomx['account.Charges.Total'].fillna(datos_telecomx['account.Charges.Total'].mean())
```

### Paso 2: Eliminar duplicados

```python
# Mostrar los valores duplicados en el Data frame
print(f"Duplicados antes: {datos_telecomx.duplicated().sum()}")
# Eliminarlos completamente del Data frame
datos_telecomx = datos_telecomx.drop_duplicates()
```

### Paso 3: Revisar y convertir tipos de datos

```python
print(datos_telecomx.dtypes)

# Convertir columnas numéricas mal cargadas como string
# datos_telecomx['account.Charges.Total'] = pd.to_numeric(datos_telecomx['account.Charges.Total'], errors='coerce')
# datos_telecomx['account.Charges.Monthly'] = pd.to_numeric(datos_telecomx['account.Charges.Monthly'], errors='coerce')
```

###  Paso 4: Limpieza de texto en columnas categóricas

```python
# Lista de columnas del Data frame
cat_cols = ['Churn', 'customer.gender', 'customer.Partner', 'customer.Dependents', 'phone.PhoneService',
            'phone.MultipleLines', 'internet.InternetService', 'internet.OnlineSecurity', 'internet.OnlineBackup',
            'internet.DeviceProtection', 'internet.TechSupport', 'internet.StreamingTV', 'internet.StreamingMovies',
            'account.Contract', 'account.PaperlessBilling', 'account.PaymentMethod']

# Recorrer cada una de las columnas de texto y poner la información en minúsculas
for col in cat_cols:
    datos_telecomx[col] = datos_telecomx[col].astype(str).str.strip().str.lower()
```

### Paso 5: Verificar categorías únicas (inconsistencias)

```python
# Recorrer cada una de las columnas de texto y validar si las categorias son únicas
for col in cat_cols:
    print(f"{col}: {datos_telecomx[col].unique()}")
```

```python
# Modicficar valores de la columna Chrun vacias usando la moda
datos_telecomx['Churn'] = datos_telecomx['Churn'].fillna(datos_telecomx['Churn'].mode()[0])
```

### Paso 6: Normalizar fechas (si hubiera fechas)

```python
# Suponiendo que hay una columna 'FechaAlta'
# df['FechaAlta'] = pd.to_datetime(df['FechaAlta'])
# df['FechaAlta'] = df['FechaAlta'].dt.normalize()
```

### Agregar la columna de cuentas diarias

```python
# Creación de la columna Cuentas_Diarias en el Data frame
datos_telecomx['Cuentas_Diarias'] = datos_telecomx['account.Charges.Monthly'] / 30

# Mostrar el Data frame con la nueva columna Cuentas_Diarias
print(datos_telecomx[['account.Charges.Monthly', 'Cuentas_Diarias']].head())
```

### Estandarización y transformación de datos

####  1. Convertir texto a binario (Sí/No → 1/0)

```python
binarias = ['Churn', 'customer.Partner', 'customer.Dependents',
            'phone.PhoneService', 'account.PaperlessBilling']

for col in binarias:
    datos_telecomx[col] = datos_telecomx[col].map({'yes': 1, 'no': 0})
```

#### 2. Reemplazar valores como “no internet service” por “no”

```python
cols_internet = ['internet.OnlineSecurity', 'internet.OnlineBackup',
                 'internet.DeviceProtection', 'internet.TechSupport',
                 'internet.StreamingTV', 'internet.StreamingMovies']

for col in cols_internet:
    datos_telecomx[col] = datos_telecomx[col].replace('no internet service', 'no')
```

#### 3. Estandarizar formatos (minúsculas, sin espacios)

```python
for col in datos_telecomx.select_dtypes(include='object').columns:
    datos_telecomx[col] = datos_telecomx[col].str.strip().str.lower()
```

#### Renombrar columnas para claridad (opcional)

```python
datos_telecomx.rename(columns={
    'account.Charges.Monthly': 'gasto_mensual',
    'account.Charges.Total': 'gasto_total',
    'Cuentas_Diarias': 'gasto_diario'
}, inplace=True)
```

## 📘 Análisis de Evasión de Clientes (Churn)

---
### 🔹 1. Introducción

El presente análisis tiene como objetivo comprender el fenómeno de **evasión de clientes (Churn)** en una empresa de telecomunicaciones.  
El abandono de clientes impacta directamente en los ingresos y la sostenibilidad del negocio, por lo que identificar patrones que lo predicen puede mejorar la retención.

Se trabajó con un conjunto de datos que incluye información sobre clientes, tipo de contrato, servicios contratados y comportamiento de pago.

**Pregunta clave**:  
*¿Qué factores están asociados a una mayor probabilidad de evasión?*

---

### 🔹 2. Limpieza y Tratamiento de Datos

- Importación de datos desde un archivo JSON alojado en GitHub.
- Revisión y eliminación de valores nulos y duplicados.
- Unificación de categorías con respuestas similares o inconsistentes.
- Conversión de variables categóricas binarias a formato numérico (1/0).
- Creación de nueva variable `Cuentas_Diarias` (gasto mensual / 30).
- Renombramiento de columnas para mejorar la legibilidad.

---
### 🔹 3. Análisis Exploratorio de Datos (EDA)

Se aplicaron técnicas estadísticas y visualizaciones para explorar los datos:

- Estadísticas descriptivas con `.describe()`.
- Visualización de la variable `Churn` con gráficos de barras y pastel.
- Comparación de churn según variables categóricas: género, tipo de contrato, método de pago.
- Análisis de variables numéricas (`gasto_total`, `tenure`) mediante boxplots e histogramas.

---

### 🔹 4. Conclusiones e Insights

- El tipo de contrato es uno de los principales indicadores de evasión.
- Clientes con menor antigüedad y bajo gasto presentan mayor riesgo de baja.
- Hay comportamientos distintos entre usuarios con servicios desactivados como “no internet service”.
- Existen perfiles con alto churn que requieren atención especial.

---
### 🔹 5. Recomendaciones Estratégicas

### Exploración Visual del Churn

#### 🔹 Exploración Visual del Churn

En esta sección analizamos visualmente la evasión de clientes (Churn) con base en diferentes variables. Los gráficos permiten identificar patrones y relaciones clave entre los atributos del cliente y su propensión a cancelar el servicio.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=datos_telecomx, x='Churn', palette='Set2', hue='Churn', legend=False)
plt.title('Distribución de Churn')
plt.xlabel('Churn')
plt.ylabel('Cantidad de Clientes')
plt.show()
```

La proporción de clientes que cancelan es significativamente mayor en contratos mensuales, lo que sugiere menor compromiso.

#### 🔹 Churn por método de pago

Los métodos de pago también parecen estar relacionados con la evasión. El método "electronic check" muestra una mayor tasa de cancelación.

#### 🔹 Distribución de Tenure por Churn

El tiempo que un cliente ha estado en la empresa también muestra diferencias claras. Los clientes que cancelan tienden a tener menor "tenure".

#### 🔹 Matriz de Correlación

La siguiente matriz muestra las correlaciones entre variables numéricas. Los valores cercanos a 1 o -1 indican una fuerte correlación positiva o negativa, respectivamente. Esta visualización puede ayudar a detectar variables influyentes en la evasión.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar solo las columnas numéricas
numericas = datos_telecomx.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
corr = numericas.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.tight_layout()
plt.show()
```

#### 🔹 Correlaciones Clave con la Evasión

Exploramos cómo algunas variables numéricas y de comportamiento se relacionan con el churn (evasión), incluyendo:

- El gasto diario (cuenta diaria) versus churn.
- La cantidad de servicios contratados versus churn.

Estas relaciones pueden ofrecer pistas para identificar factores de riesgo en la cancelación de clientes.

##### Correlación entre Cuentas_Diarias y Churn

```python
# Convertir la variable Churn a binaria para análisis de correlación
datos_telecomx['Churn_bin'] = datos_telecomx['Churn'].map({'yes': 1, 'no': 0})

# Gráfico de dispersión entre Cuentas_Diarias y Churn
sns.boxplot(data=datos_telecomx, x='Churn', y='gasto_diario', palette='Set2', hue='Churn', legend=False)
plt.title('Relación entre Gasto Diario y Evasión')
plt.xlabel('Churn')
plt.ylabel('Cuentas Diarias')
plt.show()

# Correlación numérica
corr_diaria = datos_telecomx[['gasto_diario', 'Churn_bin']].corr().iloc[0,1]
print(f'Correlación entre cuenta diaria y churn: {corr_diaria:.2f}')
```

##### Relación entre cantidad de servicios y churn

```python
# Crear columna con cantidad de servicios activos
servicios = [
    'phone.MultipleLines', 'internet.OnlineSecurity', 'internet.OnlineBackup',
    'internet.DeviceProtection', 'internet.TechSupport',
    'internet.StreamingTV', 'internet.StreamingMovies'
]

# Contar cuántos servicios adicionales tiene el cliente
datos_telecomx['Servicios_Activos'] = datos_telecomx[servicios].apply(
    lambda fila: sum(fila == 'yes'), axis=1)

# Visualizar relación con churn
sns.boxplot(data=datos_telecomx, x='Churn', y='Servicios_Activos', palette='coolwarm', hue='Churn', legend=False)
plt.title('Cantidad de Servicios vs. Churn')
plt.xlabel('Churn')
plt.ylabel('Servicios Activos')
plt.show()

# Correlación numérica
corr_servicios = datos_telecomx[['Servicios_Activos', 'Churn_bin']].corr().iloc[0,1]
print(f'Correlación entre servicios activos y churn: {corr_servicios:.2f}')
```
